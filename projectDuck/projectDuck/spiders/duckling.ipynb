```python
import io
import pandas
import json
from collections import defaultdict
# from collections import OrderedDict, defaultdict
```

```python
with open(r"C:\Users\APPKVc1000\projectDuck\duck.json"):
    domains = [json.loads(domain) for domain in open(
        r"C:\Users\APPKVc1000\projectDuck\duck.json",
        encoding='utf-8').readlines()]
```

```python
kingdom = defaultdict(list)

for domain in domains:
    for webpage, data in domain.items():
        if type(data) == str:
            kingdom[webpage].append(data)

kingdom = dict(kingdom)
```

```python
phylum = dict()

for domain in domains:
    phyla = defaultdict(list)
    {phyla[webpages].append(url) for webpages, url in domain.items() if isinstance(url, dict)}
    phyla = dict(phyla)
    for website, data in phyla.items():
        for webpage in data:
            for url, classes in webpage.items():
                phylum[(website, url)] = classes
```

```python
order = list()

for webpage, classes in phylum.items():
    for class_data, data in classes.items():
        order.append([duck for duck in data])

families = list()
genera = list()
species = list()

for duck in order:
    for data in duck:
        if isinstance(data, dict):
            for family, genus in data.items():
                families.append(family)
        elif isinstance(data, list):
            for genus in data:
                genera.append(genus)
        elif isinstance(data, str):
            species.append(data)
        else:
            raise ValueError
```
